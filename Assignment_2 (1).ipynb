```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('glass.csv')
```

```python
df["Window"]=df["Type"].map({1:0,2:0,3:0,4:0,5:1,6:1,7:1})
```

```python
X=df.iloc[:,:-2].values
y=df["Window"].values.reshape(-1,1)
print(X)
```

```python
class SLPNN:
    def __init__(self, alpha,epochs):
        self.lr = alpha
        self.epochs = epochs


    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def feedforward(self, X):
        z = np.dot(X, self.w)
        h = self.sigmoid(z)
        return h

    def error(self, h, y):
        e = (-y * np.log(h + 1e-10) - (1 - y) * np.log(1 - h + 1e-10)).mean()
        self.E.append(e)

    def backprop(self, X, y, h):
        delta = np.dot(X.T, (h - y)) / len(X)
        self.w = self.w - self.lr * delta

    def train(self, X, y):
      self.w=np.zeros((X.shape[1],1))
      self.E=[]
      for epoch in range(self.epochs):
            h = self.feedforward(X)
            self.backprop(X, y, h)
            self.error(h, y)

            if (epoch + 1) % 1000 == 0:
                print("Epoch:", epoch + 1, "Error:", round(self.E[-1], 4))

    def predict(self, X):
        return self.feedforward(X)


    def classify(self, X):
        return np.round(self.predict(X))
```

```python
model = SLPNN(alpha=0.01,epochs=1000)
model.train(X, y)
pred=model.predict(X)
pred2=model.classify(X)
print("Probabilities predicted:",pred[:10])
print("Class predicted:",pred2[:10])
```

```python
accuracy=np.mean(pred2==y)
print("Accuracy:",accuracy)
```
